```python
## test readout h5 file from recorder
fp = "/masc_shared/ag_volz/belz_from_haust/10172024_AG_Gottfried_3010"
fn = 'GridE1_NBD25nm_1350nmx1350nm_100x100x3x4nmx3x4nm_CL150cm_400ms_3kB8__20241017_112232_converted_less_lacy_carbon_area_[3, 48, 31, 99].nc'

fn_save = 'GridE1_NBD25nm_1350nmx1350nm_100x100x3x4nmx3x4nm_CL150cm_400ms_3kB8__20241017_112232_converted_less_lacy_carbon_area_[3, 48, 31, 99].h5'

EELS_File= h5py.File(os.path.join(fp,fn),'r') 

```

```python
EELS_MAP = EELS_File['corrected_data/remove_dead_pixel/data_4d']
EELS_MAP = hs.signals.Signal2D(np.asarray(EELS_MAP).astype(float))
```

```python
## save path

datacube_save_path = os.path.join(fp,fn_save)

datacube = py4DSTEM.datacube.datacube.DataCube(
                    data=EELS_MAP.data,
                    name='Dataset_h5',
                    # units= "Intensity [a.u.]",
                    # dims=None, #list
                    # dim_names= ["Rx","Ry","Qx","Qy"], #list
                    # dim_units= ["Å","Å","1/Å","1/Å"], #list
                    slicelabels=None,
                )

## save in py4dstem dataformat
py4DSTEM.save(
    datacube_save_path,
    datacube,
    mode='o'    # write' mode
)
```
