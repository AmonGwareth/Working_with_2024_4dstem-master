```python
## use pyxem test data
import pyxem as pxm
import hyperspy.api as hs
from FunctionPackageOM import BackgroundFilterFunctions
%matplotlib ipympl
```

```python
## Test PreFilterDifferenceOfGaussians 
diffractionSignal = pxm.data.tilt_boundary_data()

diffractionSignalFiltered = BackgroundFilterFunctions.PreFilterDifferenceOfGaussians(diffractionSignal.inav[2,2], min_sigma=3, max_sigma=20)
diffractionSignalFilteredSignal = hs.signals.Signal2D(diffractionSignalFiltered) #convert to signal for plotting function


hs.plot.plot_images(
    [diffractionSignal.inav[2,2], diffractionSignalFilteredSignal],
    label=["diffractionSignal", "diffractionSignalFilteredSignal"],
    tight_layout=True,
    norm="symlog",
    cmap="viridis",
    colorbar=None,
)
```
