```python
## MAG list - FOV in nm2

MAG_2_5_M_FOV_nm = 51.205**2
MAG_4_M_FOV_nm = 35.097**2
MAG_5_M_FOV_nm = 28.738**2
MAG_12_M_FOV_nm = 12.013**2
MAG_20_M_FOV_nm = 7.372**2

## make array

MAG_Array = np.array([[2.5, MAG_2_5_M_FOV_nm],[4, MAG_4_M_FOV_nm],[5, MAG_5_M_FOV_nm],[12, MAG_12_M_FOV_nm],[20, MAG_20_M_FOV_nm]])
MAG_Array
```
